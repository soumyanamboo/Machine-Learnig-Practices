#Task 2.2:
1. Introduce noise (standard deviation of the gaussian noise= 10) in the dataset created in 2.1

2. Visualize the dataset


```python
# data = make_regression(n_samples = 500,
#                        n_features = 2,
#                        noise = 10)
```

```python
import numpy as np 
# ?np.random.randn

# (sigma = 10)
noise_featurematrix = np.random.randn(500, 2)*10
noise_target = np.random.randn(500,)*10

X_noisy = X + noise_featurematrix
y_noisy = data[1] +noise_target
```

```python
plt.hist(y_noisy, bins = 20, width = 8)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(X_noisy[:,0], 
            y_noisy, 
            alpha = 0.5,
            c = 'r',
            label = 'Noisy data')
plt.scatter(X[:, 0], 
            data[1], 
            alpha = 0.5,
            c = 'b', 
            label ='Original Data')
plt.legend()
plt.show()
```

#Task 3:

1. Generate a classification data set using make_classification with the following details:
  * 2 features
  * 2 classes
  * one cluster per class



```python
from sklearn.datasets import make_classification

data = make_classification(n_features = 2,
                           n_samples = 1000, 
                           n_informative = 2,
                           n_repeated = 0, 
                           n_redundant = 0 ,
                           n_classes = 2, 
                           n_clusters_per_class=2)
```

```python
data[0].shape
```

```python
plt.scatter(data[0][: ,0], data[0][: ,1], c = data[1])
```

# Task 8: Create polynomial regression dataset

```python
def f(x):
  return x**3 + 2*x**2 - 3*x-2

x = np.linspace(-5,5,200)
y = f(x)

noise = 8*np.random.randn(200) +5
y_noisy = y + noise
plt.plot(x,y)
plt.plot(x,y_noisy)
```
